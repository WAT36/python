```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#重み
w0=-2
w1=1
w2=2
```

```python
#入力総和
def a(x0,x1):
    return w0*x0 + w1*x1 + w2

#出力値
def y(x0,x1):
    return 1/(1+np.exp(a(x0,x1)))
```

```python
fig=plt.figure()
ax = Axes3D(fig)

#ラベル指定
ax.set_xlabel("x0")
ax.set_ylabel("x1")
ax.set_zlabel("y")

#x0,x1の座標(-10~10)
x0 = np.linspace(-10,10,100)
x1 = np.linspace(-10,10,100)

#直線(w0x0+w1x1+w2=0)の座標,初期値0
x1_line=np.linspace(0,0,100)
z_line=np.linspace(0,0,100)

#直線の座標の計算
for i in range(len(x1_line)):
    x1_line[i]= (-w0*x0[i]-w2)/w1
    z_line[i]=y(x0[i],x1_line[i])

#出力値yの平面の座標,初期値0
plane=np.zeros((len(x0),len(x1)))

#平面の座標の計算
for x0i in range(len(x0)):
     for x1i in range(len(x1)):
             plane[x1i,x0i] = y(x0[x0i],x1[x1i])

x00,x11 = np.meshgrid(x0,x1)

#平面を図示（薄赤）
ax.plot_surface(x00,x11,plane,rstride=1,cstride=1,alpha=0.3,color='white',edgecolor='lightcoral')

#直線を図示（黒）
ax.plot(x0, x1_line, z_line, color="black")

#図示範囲の指定
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.view_init(elev=60, azim=135)

#表示
plt.show()
```
